## Sebességek számítása

A középiskolai fizikát felhasználva egyszerűen számíthatjuk a sebességeket a megtett távolság és az eközben eltelt idő alapján. Ehhez a számítháshoz használhajuk a pandas shift függvényét, amivel az adatainkat tudjuk a függvényben megadott létésben eltöltni. A sebesség nagyságát ezek után numpy segítségével számítjuk és létrehozunk egy új oszlopot, hogy eltároljuk. 

```python
import numpy as np

df["p_x_d"] = (df["p_x"] - df["p_x"].shift(1)) / (df["t"] - df["t"].shift(1))
df["p_y_d"] = (df["p_y"] - df["p_y"].shift(1)) / (df["t"] - df["t"].shift(1))
df["p_z_d"] = (df["p_z"] - df["p_z"].shift(1)) / (df["t"] - df["t"].shift(1))
df["p_v_abs"] = np.sqrt(
    df["p_x_d"].values ** 2 + df["p_y_d"].values ** 2 + df["p_z_d"].values ** 2
)

# df["v_x"] = (df["p_x"] - df["p_x"].shift(1)) / (df["t"] - df["t"].shift(1))
# df["v_y"] = (df["p_y"] - df["p_y"].shift(1)) / (df["t"] - df["t"].shift(1))
# df["v_z"] = (df["p_z"] - df["p_z"].shift(1)) / (df["t"] - df["t"].shift(1))
# df["p_v_abs"] = np.sqrt(
#     df["v_x"].values ** 2 + df["v_y"].values ** 2 + df["v_z"].values ** 2
# )

df
```

Ezek után meg is jelentíthetjük az abszolut segességet, mint színt a 3D-s ábránkon. 

```python
fig = px.scatter_3d(df.iloc[1:, :], x="p_x", y="p_y", z="p_z", color="p_v_abs")
fig.show()
```

A szemfülesek észrevehették, hogy a mérés tartalmaz sebesség adatokat is, nézzük meg, hogy mekkor az áltaunk számított és a mért közt a különbség.

```python
fig = px.scatter(df.iloc[1:, :], x="p_x_d", y="v_x")
fig.show()
```

Nem tűnik olyan rossznak amit csináltunk, de ha belenagyítunk, akkor azért látszik, hogy van eltérés.

## Feladat 1. 

Határozzuk meg a gyorsulásokat, és jelenítsük meg az abszolút gyorsulást egy, a sebességet ábrázolóhoz hasonló 3D-s ábrán!

```python
### SAJÁT KÓDOD ###
df["a_x"] = (df["v_x"] - df["v_x"].shift(1)) / (df["t"] - df["t"].shift(1))
df["a_y"] = (df["v_y"] - df["v_y"].shift(1)) / (df["t"] - df["t"].shift(1))
df["a_z"] = (df["v_z"] - df["v_z"].shift(1)) / (df["t"] - df["t"].shift(1))
df["a_abs"] = np.sqrt(
    df["a_x"].values ** 2 + df["a_y"].values ** 2 + df["a_z"].values ** 2
)
df
```

```python
fig = px.scatter_3d(df.iloc[1:, :], x="p_x", y="p_y", z="p_z", color="a_abs")
fig.show()
```

## Feladat 2. 

Határossuk meg a görbék torzióját és ábrázoljuk ezt is 3D ábán (a görbület skalárfüggvényét színként ábrázolva a görbén)!

```python
### SAJÁT KÓDOD ###
df["j_x"] = (df["a_x"] - df["a_x"].shift(1)) / (df["t"] - df["t"].shift(1))
df["j_y"] = (df["a_y"] - df["a_y"].shift(1)) / (df["t"] - df["t"].shift(1))
df["j_z"] = (df["a_z"] - df["a_z"].shift(1)) / (df["t"] - df["t"].shift(1))
```

```python
df["tau"] = (
      df["j_x"] * (df["v_y"] * df["a_z"] - df["v_z"] * df["a_y"])
    + df["j_y"] * (df["v_z"] * df["a_x"] - df["v_x"] * df["a_z"])
    + df["j_z"] * (df["v_x"] * df["a_y"] - df["v_y"] * df["a_x"])
) / ( 
      (df["v_y"] * df["a_z"] - df["v_z"] * df["a_y"]) ** 2
    + (df["v_z"] * df["a_x"] - df["v_x"] * df["a_z"]) ** 2
    + (df["v_x"] * df["a_y"] - df["v_y"] * df["a_x"]) ** 2
)
```

```python
fig = px.scatter_3d(
    df.iloc[1:, :], x="p_x", y="p_y", z="p_z", color="tau", range_color=(-0.5, 0.5)
)
fig.show()
```

## Feladat 3. 

Készítsünk egy függvényt, ami megadja a görbe adott pontjának kisérő triéder vektorait!


```python
def kisero_trieder(df=df, pont=100):
    # dataframe-ben vannak megadott görbe pontjai
    # a pont azt jelöli ki, hogy hanyadik görbe pont környezetében vagyunk kíváncsiak a kisérő triéderre.
    ### SAJÁT KÓDOD ###

    t = [df["v_x"][pont], df["v_y"][pont], df["v_z"][pont]]
    t /= np.sqrt(df["v_x"][pont] ** 2 + df["v_y"][pont] ** 2 + df["v_z"][pont] ** 2)

    b = [
        (df["v_y"][pont] * df["a_z"][pont] - df["v_z"][pont] * df["a_y"][pont]),
        (df["v_z"][pont] * df["a_x"][pont] - df["v_x"][pont] * df["a_z"][pont]),
        (df["v_x"][pont] * df["a_y"][pont] - df["v_y"][pont] * df["a_x"][pont]),
    ]
    b /= np.sqrt(
        (df["v_y"][pont] * df["a_z"][pont] - df["v_z"][pont] * df["a_y"][pont]) ** 2
        + (df["v_z"][pont] * df["a_x"][pont] - df["v_x"][pont] * df["a_z"][pont]) ** 2
        + (df["v_x"][pont] * df["a_y"][pont] - df["v_y"][pont] * df["a_x"][pont]) ** 2
    )
    n = np.cross(b, t)

    return (t, n, b)


kisero_trieder()
```
